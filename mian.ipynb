```python
selected_features = [
    'Age',
    'Gender',
    'Family_History',
    'Radiation_Exposure',
    'Iodine_Deficiency',
    'Smoking',
    'Obesity',
    'Diabetes',
    'TSH_Level',
    'T3_Level',
    'T4_Level',
    'Nodule_Size',
    'Thyroid_Cancer_Risk',
]
```

```python
selected_columns = selected_features + ['Diagnosis']
df_selected = df[selected_columns]
df_selected
```

```python
df_selected.isnull().sum()
```

```python
cat_col = df_selected.select_dtypes(include='object')
cat_col
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df_selected.copy()

le = LabelEncoder()
for col in cat_col:
    df_encoded[col] = le.fit_transform(df_encoded[col])
```

```python
df_encoded
```

```python
x = df_encoded.drop('Diagnosis', axis=1)
y = df_encoded['Diagnosis']
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
X_scaled = pd.DataFrame(x_scaled, columns=x.columns)
X_scaled
```

```python
y
```

```python
y.value_counts()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt

sns.countplot(x=y)
plt.title("Distribution of Diagnosis")
plt.xlabel("Class (Diagnosis)")
plt.ylabel("Count")
plt.show()
```

```python
df_viz = X_scaled.copy()
df_viz['Diagnosis'] = y.values

corr_matrix = df_viz.corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(x_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=y, palette='Set1')
plt.title("PCA Projection of Data")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.legend(title="Diagnosis")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.naive_bayes import GaussianNB

NB = GaussianNB()
NB.fit(x_train, y_train)

y_pred_train_NB = NB.predict(x_train)
y_pred_test_NB = NB.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, classification_report

acc_train = accuracy_score(y_train, y_pred_train_NB)
acc_test = accuracy_score(y_test, y_pred_test_NB)

acc_train, acc_test
```

```python
cm = confusion_matrix(y_test, y_pred_test_NB)
print(cm)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title(f"Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors=4)
KNN.fit(x_train, y_train)

y_pred_test_KNN = KNN.predict(x_test)
y_pred_train_KNN = KNN.predict(x_train)
```

```python
acc_train = accuracy_score(y_train, y_pred_train_KNN)
acc_test = accuracy_score(y_test, y_pred_test_KNN)

acc_train, acc_test
```

```python
cm = confusion_matrix(y_test, y_pred_test_KNN)
print(cm)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title(f"Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(n_estimators=100)
RF.fit(x_train, y_train)
y_pred_test_RF = RF.predict(x_test)
y_pred_train_RF = RF.predict(x_train)
```

```python
acc_test = accuracy_score(y_pred_test_RF, y_test)
acc_train = accuracy_score(y_pred_train_RF, y_train)

acc_test, acc_train
```

```python
cm = confusion_matrix(y_test, y_pred_test_RF)
print(cm)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title(f"Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
def evaluate_model(name, y_test, y_pred):
    print(f"\n=== {name} ===")
    acc = accuracy_score(y_test, y_pred)*100
    print(f"Accuracy: {acc:.4f}\n")
    
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
```

```python
evaluate_model("K-Nearest Neighbors", y_test, y_pred_test_KNN)
evaluate_model("Random Forest", y_test, y_pred_test_RF)
evaluate_model("Naive Bayes ", y_test, y_pred_test_NB)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
def final_report(models, X_test, y_test):
    
    plt.style.use('_mpl-gallery')

    sns.set(style="whitegrid")
    plt.figure(figsize=(12,8))

    auc_scores = {}
    accuracies = {}
    
    for name, model in models.items():
        y_prob = model.predict_proba(X_test)[:,1]
        y_pred = model.predict(X_test)
        
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        auc = roc_auc_score(y_test, y_prob)
        acc = accuracy_score(y_test, y_pred)
        
        auc_scores[name] = auc
        accuracies[name] = acc
        
        plt.plot(fpr, tpr, linewidth=3, label=f"{name} (AUC={auc:.3f}, Acc={acc:.3f})")
    
    plt.plot([0,1], [0,1], 'k--', linewidth=1.5)
    
    plt.xlabel('False Positive Rate', fontsize=14)
    plt.ylabel('True Positive Rate', fontsize=14)
    plt.title(' ROC Curve Comparison & Model Performance', fontsize=16, fontweight='bold')
    plt.legend(loc='lower right', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
    
    print(" Final Model Analysis:\n")
    sorted_auc = sorted(auc_scores.items(), key=lambda x: x[1], reverse=True)
    sorted_acc = sorted(accuracies.items(), key=lambda x: x[1], reverse=True)

    best_auc_model, best_auc = sorted_auc[0]
    best_acc_model, best_acc = sorted_acc[0]

    print(f" Best model based on AUC: {best_auc_model} with a score of {best_auc:.3f}")
    print(f" Best model based on Accuracy: {best_acc_model} with a score of {best_acc:.3f}\n")

    if best_auc_model == best_acc_model:
        print(f"Conclusion: The model {best_auc_model} performs best overall, achieving the highest AUC and Accuracy.")
    else:
        print("Conclusion: Models differ in terms of AUC and Accuracy;")
        print(f" - {best_auc_model} has the highest AUC, indicating better discrimination capability.")
        print(f" - {best_acc_model} has the highest Accuracy, indicating more correct predictions on the test set.")
        print("You should select the model depending on which metric is more important for your task.")

    print("\n Additional Notes:")
    print("- AUC measures how well the model distinguishes between classes and is especially useful with imbalanced datasets.")

    print("- Accuracy indicates the overall percentage of correct predictions.")
    print("- The ROC curve helps visually compare the performance of different models.\n")

    print(" Suggestion: For further improvements, consider hyperparameter tuning and feature selection.")
```

```python
models_dict = {
    "K-Nearest Neighbors": KNN,
    "Random Forest": RF,
    "Gradient Boosting": NB
}

final_report(models_dict, x_test, y_test)
```
